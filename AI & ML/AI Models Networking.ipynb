# 🔥 3️⃣ REINFORCEMENT LEARNING

## 🎯 Use Case: Dynamic Path Selection

Instead of static OSPF cost.

---

## 🧠 Step 1 — Define Environment

State:

* Link utilization level

Action:

* Choose Path A or B

Reward:

* Lower congestion = higher reward

---

## 🧠 Step 2 — Q-Learning Example

```python
import numpy as np
import random

q_table = np.zeros((3, 2))  # 3 congestion states, 2 paths

alpha = 0.8
gamma = 0.9
epsilon = 0.1

def get_state(util):
    if util < 40:
        return 0
    elif util < 70:
        return 1
    else:
        return 2

for episode in range(1000):
    utilization = random.randint(0, 100)
    state = get_state(utilization)

    if random.random() < epsilon:
        action = random.choice([0,1])
    else:
        action = np.argmax(q_table[state])

    if state == 2 and action == 0:
        reward = -10
    else:
        reward = 10

    next_state = get_state(random.randint(0,100))

    q_table[state, action] += alpha * (
        reward + gamma*np.max(q_table[next_state]) - q_table[state, action]
    )

print("Learned Routing Policy:")
print(q_table)
```

---

## 🧠 Step 3 — Apply Decision to Network

```python
best_action = np.argmax(q_table[state])

    new_cost = 50

    f"interface Gig0/0",
    f"ip ospf cost {new_cost}"
```

🧠 Step 1 — Traffic Dataset

````python
import numpy as np
from sklearn.ensemble import IsolationForest

# Normal traffic
normal = np.random.normal(1000, 50, (100, 2))

# Inject anomaly
anomaly = np.array([[5000, 6000]])

data = np.vstack([normal, anomaly])
````

🧠 Step 2 — Train Isolation Forest

````python
model = IsolationForest(contamination=0.01)
model.fit(data)

predictions = model.predict(data)

print(predictions[-1])  # -1 means anomaly
````

🧠 Step 3 — Automate Mitigation

````python
if predictions[-1] == -1:
    print("🚨 Anomaly detected — pushing ACL")

    config_commands = [
        "ip access-list extended BLOCK_DDOS",
        "deny ip any any",
        "interface Gig0/0",
        "ip access-group BLOCK_DDOS in"
    ]

    connection.send_config_set(config_commands)
````

Step 2 — Q-Learning Example

````python
import numpy as np
import random

q_table = np.zeros((3, 2))  # 3 congestion states, 2 paths

alpha = 0.8
gamma = 0.9
epsilon = 0.1

def get_state(util):
    if util < 40:
        return 0
    elif util < 70:
        return 1
    else:
        return 2

for episode in range(1000):
    utilization = random.randint(0, 100)
    state = get_state(utilization)

    if random.random() < epsilon:
        action = random.choice([0,1])
    else:
        action = np.argmax(q_table[state])

    if state == 2 and action == 0:
        reward = -10
    else:
        reward = 10

    next_state = get_state(random.randint(0,100))

    q_table[state, action] += alpha * (
        reward + gamma*np.max(q_table[next_state]) - q_table[state, action]
    )

print("Learned Routing Policy:")
print(q_table)
````

Step 3 — Apply Decision to Network

````python
best_action = np.argmax(q_table[state])

if best_action == 0:
    new_cost = 10
else:
    new_cost = 50

config_commands = [
    f"interface Gig0/0",
    f"ip ospf cost {new_cost}"
]

connection.send_config_set(config_commands)
````

Q-Learning Example

````python
import numpy as np
import random

q_table = np.zeros((3,2))  # 3 states, 2 ISPs

alpha = 0.8
gamma = 0.9
epsilon = 0.1

def get_state(latency):
    if latency < 20:
        return 0
    elif latency < 50:
        return 1
    else:
        return 2

for episode in range(1000):
    latency = random.randint(5,100)
    state = get_state(latency)

    if random.random() < epsilon:
        action = random.choice([0,1])
    else:
        action = np.argmax(q_table[state])

    if action == 0:
        reward = 100 - latency
    else:
        reward = 80 - latency

    next_state = get_state(random.randint(5,100))

    q_table[state, action] += alpha * (
        reward + gamma*np.max(q_table[next_state]) - q_table[state, action]
    )

print("Optimized BGP Policy:")
print(q_table)
````

Apply Best ISP Decision

````python
current_latency = 45
state = get_state(current_latency)
best_isp = np.argmax(q_table[state])

if best_isp == 0:
    print("Choosing ISP1")
else:
    print("Choosing ISP2")
````
