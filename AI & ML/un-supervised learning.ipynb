Here are common **Unsupervised Learning** models with Python examples using `scikit-learn`.

Unsupervised learning means **no labeled output (y)** — the model tries to find patterns, clusters, or structure in data.

---

# 🔹 1. K-Means Clustering

Used to group data into K clusters.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Create model
model = KMeans(n_clusters=3, random_state=42)

# Train model
model.fit(X)

# Predict cluster labels
labels = model.predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("K-Means Clustering")
plt.show()
```

---

# 🔹 2. Hierarchical Clustering (Agglomerative)

Builds clusters step-by-step (bottom-up).

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate data
X, _ = make_blobs(n_samples=200, centers=3, random_state=42)

# Create model
model = AgglomerativeClustering(n_clusters=3)

# Fit and predict
labels = model.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Hierarchical Clustering")
plt.show()
```

# 🔹 3. DBSCAN (Density-Based Clustering)

Groups points based on density (good for noise detection).

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate non-linear dataset
X, _ = make_moons(n_samples=300, noise=0.05)

# Create model
model = DBSCAN(eps=0.3, min_samples=5)

# Fit and predict
labels = model.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("DBSCAN Clustering")
plt.show()
```

# 🔹 4. Principal Component Analysis (PCA)

Used for dimensionality reduction.

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load data
data = load_iris()
X = data.data

# Create PCA model
pca = PCA(n_components=2)

# Transform data
X_reduced = pca.fit_transform(X)

# Plot
plt.scatter(X_reduced[:, 0], X_reduced[:, 1])
plt.title("PCA Dimensionality Reduction")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

# 🔹 6. Autoencoder (Neural Network Based)

Used for dimensionality reduction or anomaly detection.

```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np

# Generate dummy data
X = np.random.rand(1000, 20)

# Define autoencoder
input_layer = layers.Input(shape=(20,))
encoded = layers.Dense(10, activation='relu')(input_layer)
decoded = layers.Dense(20, activation='sigmoid')(encoded)

autoencoder = tf.keras.Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# Train
autoencoder.fit(X, X, epochs=10, batch_size=32)

# Encoded representation
encoder = tf.keras.Model(input_layer, encoded)
encoded_data = encoder.predict(X)
```

. Isolation Forest (Anomaly Detection)

Used for detecting anomalies in data.

````python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

# Generate normal data
X = 0.3 * np.random.randn(100, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
X = np.vstack([X, X_outliers])

# Create model
model = IsolationForest(contamination=0.1)

# Fit and predict
labels = model.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Isolation Forest Anomaly Detection")
plt.show()
````

````python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Example network telemetry dataset
data = pd.DataFrame({
    "in_bps": [200, 220, 210, 205, 5000],
    "out_bps": [180, 190, 175, 200, 6000],
    "cpu": [30, 35, 32, 29, 95]
})

model = IsolationForest(contamination=0.2)
data["anomaly"] = model.fit_predict(data)

print(data)
````

````python
from sklearn.cluster import KMeans
import pandas as pd

data = pd.DataFrame({
    "interfaces": [48, 48, 8, 8],
    "bgp_peers": [2, 3, 20, 18],
    "traffic_gbps": [10, 12, 40, 38]
})

kmeans = KMeans(n_clusters=2)
data["cluster"] = kmeans.fit_predict(data)

print(data)
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

logs = [
    "BGP neighbor down",
    "BGP neighbor reset",
    "Interface Gi0/1 down",
    "Interface Gi0/1 up"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(logs)

kmeans = KMeans(n_clusters=2)
labels = kmeans.fit_predict(X)

print(labels)
````
