## ✅ Code Explanation (Line by Line)

### **1. Import libraries**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

* **seaborn (sns)** → used for advanced statistical plotting (built on matplotlib).
* **matplotlib.pyplot (plt)** → used to display plots and control figure behavior.
* **pandas (pd)** → used for handling tabular data (DataFrame).
* **numpy (np)** → used for numerical operations and generating random data.

---

### **2. Generate random data**

```python
data = np.random.rand(100,4)
```

* `np.random.rand(100,4)` generates **random floating-point numbers** between **0 and 1**
* Shape: **100 rows and 4 columns**

Example output (concept):

```
[[0.23, 0.78, 0.44, 0.12],
 [0.56, 0.10, 0.99, 0.33],
 ...]
```

So this simulates a dataset with **100 samples** and **4 features**.

---

### **3. Convert to Pandas DataFrame**

```python
df = pd.DataFrame(data, columns=['A','B','C','D'])
```

* Converts the numpy array into a structured DataFrame.
* Column names are set as: **A, B, C, D**

Now df looks like:

| A    | B    | C    | D    |
| ---- | ---- | ---- | ---- |
| 0.23 | 0.78 | 0.44 | 0.12 |
| 0.56 | 0.10 | 0.99 | 0.33 |

---

### **4. Create Pairplot**

```python
sns.pairplot(df)
```

This is the key part.

**pairplot()** creates:

* Scatter plots between every pair of columns
* Histogram (or KDE plot) on the diagonal

So it will plot:

* A vs B
* A vs C
* A vs D
* B vs C
* B vs D
* C vs D
* plus histograms for A, B, C, D

---

### **5. Display the plot**

```python
plt.show()
```

* Displays the generated plots in a window or notebook output.

---

## ✅ What is the Use of This Code?

### **Main Use: Exploratory Data Analysis (EDA)**

Pairplot is used when you want to quickly check:

### 🔹 1. Relationship between variables

Example: Does A increase when B increases?

### 🔹 2. Correlation patterns

If the scatter plot looks like a straight line, it indicates correlation.

### 🔹 3. Distribution of each column

Diagonal histograms show whether the feature is:

* normally distributed
* skewed
* uniform
* contains outliers

### 🔹 4. Detect outliers

Points far away from the cluster are easy to spot.

---

## 🔥 In Real World Example (Networking)

Imagine a dataset like:

| latency | jitter | packet_loss | throughput |
| ------- | ------ | ----------- | ---------- |

A pairplot helps you visually check:

* Does **packet_loss increase** when **latency increases**?
* Does **throughput decrease** when **jitter increases**?

That’s why it’s very useful in performance troubleshooting.

# ✅ 1. Scatter Plot (Relationship between 2 variables)

### Example: Latency vs Packet Loss

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "latency": [10, 20, 30, 40, 50],

```

📌 Used to check correlation between two values.

# ✅ 2. Line Plot (Trend over time)

### Example: Network traffic over time

```python

    "traffic_mbps": [100, 120, 150, 130, 170]
})
```

## 1️⃣ **Lineplot (Time Series)**

* **Use:** To show trends over time.
* **Networking example:** Network traffic (Mbps) over a day, latency over hours, CPU usage over time.
* **Why:** Helps identify patterns, spikes, or drops in performance.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample network traffic data over 12 hours
df = pd.DataFrame({
    "hour": range(1,13),
    "traffic_mbps": [50, 55, 60, 70, 65, 80, 90, 85, 75, 60, 55, 50]
})

sns.lineplot(x="hour", y="traffic_mbps", data=df, marker="o")
plt.title("Network Traffic Over Time")
plt.xlabel("Hour")
plt.ylabel("Traffic (Mbps)")
plt.show()
```

---

## 2️⃣ **Scatterplot (Correlation)**

* **Use:** To visualize the relationship between **two variables**.
* **Networking example:** Latency vs Packet Loss, or Throughput vs Jitter.
* **Why:** Helps see correlation or trends between two metrics.

```python
df = pd.DataFrame({
    "latency_ms": [10, 20, 30, 40, 50, 60],
    "packet_loss_pct": [0, 1, 1.5, 2, 3, 5]
})

sns.scatterplot(x="latency_ms", y="packet_loss_pct", data=df)
plt.title("Latency vs Packet Loss")
plt.show()
```

---

## 3️⃣ **Heatmap (Correlation Matrix)**

* **Use:** To check correlation between **multiple metrics at once**.
* **Networking example:** Latency, Jitter, Packet Loss, Throughput.
* **Why:** Quickly see which metrics are strongly correlated.

```python
import numpy as np

# Random sample network metrics
df = pd.DataFrame(np.random.rand(50,4), columns=["Latency","Jitter","PacketLoss","Throughput"])
corr = df.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Network Metrics")
plt.show()
```

---

## 4️⃣ **Boxplot / Violinplot (Outliers & Distribution)**

* **Use:** To see **spread, median, and outliers** in data.
* **Networking example:** Latency per device, Packet Loss per ISP.
* **Why:** Spot spikes or unusual values quickly.

```python
# Latency per device
df = pd.DataFrame({
    "device": ["R1","R1","R2","R2","R3","R3"],
    "latency_ms": [20, 22, 30, 35, 25, 80]  # 80 is an outlier
})

sns.boxplot(x="device", y="latency_ms", data=df)
plt.title("Latency per Device")
plt.show()
```

* **Violinplot** shows the same but also the **distribution shape**.

```python
sns.violinplot(x="device", y="latency_ms", data=df)
plt.title("Latency Distribution per Device")
plt.show()
```

---

## 5️⃣ **Histogram (Distribution)**

* **Use:** To see how values are **distributed**.
* **Networking example:** Distribution of latency values across all devices.
* **Why:** Check if values are mostly low, high, uniform, or skewed.

```python
latency = [10, 12, 15, 20, 20, 22, 25, 30, 35, 50, 60]  # sample latency

sns.histplot(latency, bins=6, kde=True)
plt.title("Latency Distribution")
plt.xlabel("Latency (ms)")
plt.ylabel("Frequency")
plt.show()
```
