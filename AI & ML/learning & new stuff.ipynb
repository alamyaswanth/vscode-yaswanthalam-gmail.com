```python
import pyvista as pv
import numpy as np

sphere = pv.Sphere(radius=1)

noise = np.random.rand(sphere.n_points)
sphere["elevation"] = noise
plotter = pv.Plotter
plotter.add_mesh(sphere, scalars="elevation", cmap="terrain")
plotter.show()
```
