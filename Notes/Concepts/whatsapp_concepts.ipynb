     Return the first duplicate in a list. 

```python
def first_duplicate(lst):
    seen = set()
    for x in lst:
        if x in seen:
            return x
        seen.add(x)
    return None

print(first_duplicate([3, 1, 3, 4, 2]))
```

```python
def duplicate(lst1):
    test = set()
    for j in lst1:
        if j in test:
            return j
        test.add(j)

print(duplicate([1,5,6,3,2,1,3,5]))
```
