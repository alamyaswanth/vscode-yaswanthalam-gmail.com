```python
def brew_tea():
    start_time = time.time()
    print("Brewing tea .......")
    time.sleep(5)
    print("tea is ready.......................")
    end_time = time.time()
    print(f"Task time: {end_time - start_time} seconds")

brew_tea()
```

```python
def timer_dec(base_fn):
    def enhanced_fn():
        start_time = time.time()
        base_fn()
        end_time = time.time()
        print(f"Task time: {end_time - start_time} seconds")
    return enhanced_fn

def brew_tea():
    print("Brewing tea .......")
    time.sleep(5)
    print("tea is ready.......................")
timer_dec(brew_tea)


```
