Python String Slicing Examples with For Loops | Tutorial

```python
message = "Yaswanth_Alam and I am"

for i in range(len(message)):
    print(message[i])
```

```python
message = "Yaswanth_Alam and I am"

for i in range(len(message)):
    print(message[i : i+1])
```

```python
message = "Yaswanth_Alam and I am"

for i in range(len(message)):
    print(message[i : i+2])
```

```python
message = "Yaswanth_Alam and I am  !"

for i in range(len(message)):
    print(message[i : i+3])
```

```python
message1 = "Yaswanth_Alam and I am !"

for i in range(len(message1)):
    print(message1[i : i+2])
```

```python
message1 = "Yaswanth_Alam and I am !"

for i in range(len(message1)):
    print(message1[i:i+2])
```

```python
message1 = "Yaswanth_Alam and I am !"

for i in range(len(message1)-1):
    print(message1[i:i+2])
```
