## 1️⃣ What is `try` and `except`?

`try`–`except` is used for **exception handling**.
It helps your program **not crash** when an error occurs.

### Basic idea

* Put **risky code** inside `try`
* Handle **errors** inside `except`

```python
try:
    # code that may cause an error
except:
    # code that runs if an error occurs
```

---

## 2️⃣ Simple Example: Division Error

### ❌ Without try-except (program crashes)

```python
x = 10
y = 0
print(x / y)
```

Output:

```
ZeroDivisionError: division by zero
```

---

### ✅ With try-except

```python
try:
    x = 10
    y = 0
    print(x / y)
except:
    print("You cannot divide by zero")
```

✔ Program continues safely.

---

## 3️⃣ Catching Specific Exceptions (Best Practice)

Instead of catching **all errors**, catch **specific ones**.

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x / y)
except ZeroDivisionError:
    print("Division by zero is not allowed")
except ValueError:
    print("Please enter only numbers")
```

### Why this is good?

* You know **exactly what went wrong**
* Easier to debug

---

## 4️⃣ Using `else` (Runs if NO error)

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Calculation successful")
```

### Flow:

* `try` → run code
* `except` → run only if error
* `else` → run only if **no error**

---

## 5️⃣ Using `finally` (Always Runs)

`finally` runs **no matter what** (error or not).

```python
try:
    f = open("test.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Closing file operation")
```

📌 Common use:

* Closing files
* Releasing resources
* Cleanup tasks

---

## 6️⃣ Handling Multiple Errors Together

```python
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    print(a / b)
except (ZeroDivisionError, ValueError):
    print("Invalid input or division by zero")
```

---

## 7️⃣ Getting the Error Message

```python
try:
    x = int("abc")
except ValueError as e:
    print("Error message:", e)
```

Output:

```
Error message: invalid literal for int() with base 10: 'abc'
```

## 8️⃣ Real-Life Example (List Index)

```python
numbers = [10, 20, 30]

try:
    print(numbers[5])
except IndexError:
    print("Index out of range")
```
