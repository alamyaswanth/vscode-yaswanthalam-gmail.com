Python follows the LEGB Rule for scope


When Python looks for a variable, it searches in this order: [realpython.com]

L – Local

Variables defined inside a function

Accessible only within that function

```python
def func():
    x = 10  # local
    print(x)

func()
# print(x) → NameError
```

E – Enclosing

Variables in an outer function, used by an inner (nested) function

```python
def outer():
    y = 20
    def inner():
        print(y)  # enclosing
    inner()
```

G – Global

Variables defined outside all functions

Accessible everywhere (read-only by default)

```python
z = 30

def func():
    print(z)

func()
```

B – Built-in


Predefined names like print, len, range

```python
print(len("Python"))
```

nonlocal keyword

Used to modify a variable from the enclosing scope (outer function), not global:

```python
def outer():
    x = 5
    def inner():
        nonlocal x
        x = 10
    inner()
    print(x)  # 10
```
