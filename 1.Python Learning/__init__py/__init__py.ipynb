In Python, **`__init__.py`** is used to **mark a directory as a package** and to **control how modules inside that package are imported**. Practicing it in **Visual Studio Code** is a great way to understand Python package structure.

I'll show **3 small practice examples** you can run in VS Code.

---

# 1️⃣ Example 1 — Basic Python Package

### Folder Structure

```
project/
│
├── main.py
└── mypackage/
    ├── __init__.py
    ├── math_utils.py
```

---

### math_utils.py

```python
def add(a, b):
    return a + b
```

---

### **init**.py

```python
print("mypackage initialized")
```

---

### main.py

```python
import mypackage.math_utils

result = mypackage.math_utils.add(5, 3)
print(result)
```

---

### Output

```
mypackage initialized
8
```

This shows that **`__init__.py` runs automatically when the package is imported.**

---

# 2️⃣ Example 2 — Export Functions from Package

Now you can **simplify imports** using `__init__.py`.

### Folder

```
project/
│
├── main.py
└── mypackage/
    ├── __init__.py
    ├── math_utils.py
```

---

### math_utils.py

```python
def multiply(a, b):
    return a * b
```

---

### **init**.py

```python
from .math_utils import multiply
```

---

### main.py

```python
from mypackage import multiply

print(multiply(4,5))
```

---

### Output

```
20
```

Here **`__init__.py` exposes functions from modules.**

---

# 3️⃣ Example 3 — Multiple Modules in a Package

### Folder Structure

```
project/
│
├── main.py
└── network/
    ├── __init__.py
    ├── ping.py
    ├── traceroute.py
```

---

### ping.py

```python
def run_ping(ip):
    print(f"Pinging {ip}")
```

---

### traceroute.py

```python
def run_trace(ip):
    print(f"Tracing route to {ip}")
```

---

### **init**.py

```python
from .ping import run_ping
from .traceroute import run_trace
```

---

### main.py

```python
from network import run_ping, run_trace

run_ping("8.8.8.8")
run_trace("8.8.8.8")
```

---

### Output

```
Pinging 8.8.8.8
Tracing route to 8.8.8.8
```

# 🧠 Key Points to Remember

`__init__.py` is used to:

1️⃣ Mark a directory as a **Python package**
2️⃣ **Run initialization code** when imported
3️⃣ **Control what gets imported**
4️⃣ Simplify imports
