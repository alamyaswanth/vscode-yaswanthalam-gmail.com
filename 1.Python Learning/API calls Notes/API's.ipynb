```python
import requests

url = "https://api.github.com/emojis"

response = requests.get(url)

print("Status Code:", response.status_code)

if response.status_code == 200:
    emojis = response.json()
    print("Total emojis:", len(emojis))

    # Print first 5 emoji names and URLs
    for name, emoji_url in list(emojis.items())[:5]:
        print(f"{name} -> {emoji_url}")
else:
    print("API call failed")
```

```python
import requests

try:
    response = requests.get("https://api.github.com/emojis", timeout=10)
    response.raise_for_status()

    emojis = response.json()
    print(f"Fetched {len(emojis)} emojis")

except requests.exceptions.RequestException as e:
    print("Error:", e)
```

```python
import requests

try:
    response = requests.get(
        "https://api.github.com/emojis",
        timeout=10
    )
    response.raise_for_status()
    print(response.json())

except requests.exceptions.Timeout:
    print("Request timed out")

except requests.exceptions.HTTPError as err:
    print("HTTP error:", err)

except Exception as e:
    print("Error:", e)
```

```python
def call_api(url, method="GET", headers=None, payload=None):
    import requests

    if method == "GET":
        return requests.get(url, headers=headers)
    elif method == "POST":
        return requests.post(url, json=payload, headers=headers)
    elif method == "PUT":
        return requests.put(url, json=payload, headers=headers)
    elif method == "DELETE":
        return requests.delete(url, headers=headers)
```

```python
import requests

# API endpoint
url = "https://jsonplaceholder.typicode.com/posts/1"

# Send GET request
response = requests.get(url)

# Check status code
if response.status_code == 200:
    data = response.json()
    print("API Response:")
    print(data)
else:
    print("Failed to fetch data:", response.status_code)
```

```python
import requests

url = "https://api.github.com/search/repositories"

params = {
    "q": "network automation",
    "sort": "stars"
}

response = requests.get(url, params=params)

data = response.json()

for repo in data["items"][:5]:
    print(repo["name"], "-", repo["html_url"])
```

```python
import requests

url = "http://192.168.1.1/api/v1/interfaces"

headers = {
    "Authorization": "Bearer mytoken123"
}

response = requests.get(url, headers=headers, verify=False)

interfaces = response.json()

for interface in interfaces:
    print(interface["name"], interface["status"])
```

```python
import subprocess

devices = ["8.8.8.8", "1.1.1.1"]

for ip in devices:

    result = subprocess.run(
        ["ping", "-c", "4", ip],
        capture_output=True,
        text=True
    )

    output = result.stdout

    for line in output.split("\n"):
        if "avg" in line:
            print(ip, line)
```

```python
import subprocess

devices = [
    "8.8.8.8",
    "1.1.1.1",
    "192.168.1.1",
    "192.168.1.10"
]

def ping_test(ip):

    result = subprocess.run(
        ["ping", "-c", "3", ip],
        capture_output=True,
        text=True
    )

    if result.returncode == 0:
        print(f"{ip} reachable")
    else:
        print(f"{ip} unreachable")

for device in devices:
    ping_test(device)
```
