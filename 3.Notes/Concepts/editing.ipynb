````python
def train_ai(X, y):
    # Learn the split point
    threshold = sum(X) / len(X)
    return threshold

def predict(number, threshold):
    return "Big" if number > threshold else "Small"

threshold = train_ai(X, y)
print("Learned threshold:", threshold)

print(predict(3, threshold))
print(predict(9, threshold))
````

Predict **whether to play outside** based on:

```python
# Features:
```
