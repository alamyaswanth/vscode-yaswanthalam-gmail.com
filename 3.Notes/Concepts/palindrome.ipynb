````python
def is_palindrome(word):
    left = 0
    right = len(word) - 1

    while left < right:
        if word[left] != word[right]:
            return False
        left += 1
        right -= 1

    return True

print(is_palindrome("civic"))
````


A **palindrome word** is a word that reads the same **forward and backward**.

### Examples

* `madam`
* `level`
* `radar`
* `civic`

---

## 1️⃣ Check if a word is a palindrome (simple)

```python
word = input("Enter a word: ")

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

### How it works

* `word[::-1]` reverses the string
* Compare original and reversed

---

## 2️⃣ Case-insensitive palindrome check

```python
word = input("Enter a word: ").lower()

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

✔ Handles `Madam`, `MADAM`, etc.

---

## 3️⃣ Using a function (best practice)

```python
def is_palindrome(word):
    return word == word[::-1]

print(is_palindrome("madam"))   # True
print(is_palindrome("hello"))   # False
```

---

## 4️⃣ Find palindrome words in a list

```python
words = ["madam", "apple", "level", "banana", "radar"]

palindromes = [w for w in words if w == w[::-1]]

print(palindromes)
```

**Output:**

```
['madam', 'level', 'radar']
```

1. `s.replace(" ", "")` — removes all spaces from the string.
2. `.lower()` — converts the string to lowercase, so the check is case-insensitive.
3. `s == s[::-1]` — compares the processed string to its reverse (`s[::-1]` reverses the string).
4. Returns `True` if the string is the same forwards and backwards (a palindrome), otherwise `False`.
