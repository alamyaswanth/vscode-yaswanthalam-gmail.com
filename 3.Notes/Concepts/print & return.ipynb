                            Return:

Return can be used only with in a function. 





```python
def func1(a,b):
    c = a + b
    return c
output = func1(2 , 6)
print(output)
```

```python
def func2(a,b):
    c = a + b
    return c
output2 = func2(2 , 6)
print(output2)
```

```python
def func1(x):
    return x+1


def func2(a,b):
    c = a + b
    return c
output2 = func2(2 , 6)
output3 = func1(output2)
print(output2)
print(output3)
```
