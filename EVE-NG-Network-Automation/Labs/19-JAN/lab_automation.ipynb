```python
from netmiko import ConnectHandler
from inventory import routers

for router in routers:
    print(f"\nConnecting to {router['name']}")

    # Create a copy without 'name' key
    connection_params = {k: v for k, v in router.items() if k != 'name'}

    conn = ConnectHandler(**connection_params)
    conn.enable()

    output = conn.send_command("show ip interface brief")
    print(output)

    conn.disconnect()
```
