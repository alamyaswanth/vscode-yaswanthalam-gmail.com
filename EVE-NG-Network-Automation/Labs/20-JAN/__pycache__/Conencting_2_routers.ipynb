````python
from netmiko import ConnectHandler

routers = [
    {
        "name": "R1",
        "device_type": "cisco_ios",
        "host": "192.168.54.142",
        "username": "admin",
        "password": "admin123",
        "secret": "admin123",
        "interface": "Ethernet1/0",
        "ip": "10.10.10.1",
        "mask": "255.255.255.252"
    },
    {
        "name": "R2",
        "device_type": "cisco_ios",
        "host": "192.168.54.143",
        "username": "admin",
        "password": "admin123",
        "secret": "admin123",
        "interface": "Ethernet1/0",
        "ip": "10.10.10.2",
        "mask": "255.255.255.252"
    }
]

for router in routers:
    print(f"\n🔧 Connecting to {router['name']} ({router['host']})")

    conn = ConnectHandler(**router)
    conn.enable()

    config_commands = [
        f"interface {router['interface']}",
        f"ip address {router['ip']} {router['mask']}",
        "no shutdown",
        "exit"
    ]

    output = conn.send_config_set(config_commands)
    print(output)

    # Save config
    conn.save_config()

    # Verify
    verify = conn.send_command("show ip interface brief")
    print("\n📊 Verification:")
    print(verify)

    conn.disconnect()
````

Great question 👍
This is a **basic but very important Netmiko automation script**, and you’re doing exactly what a real network engineer does in labs and production (just more carefully in prod 😄).

I’ll explain it **step by step**, both from a **Python point of view** and a **networking point of view**.

---

## 1️⃣ Importing Netmiko

```python
from netmiko import ConnectHandler
```

* Imports **Netmiko**, a Python library used to:

  * SSH into network devices (Cisco, Juniper, Arista, etc.)
  * Send CLI commands programmatically
* `ConnectHandler` is the main function that creates an SSH session.

---

## 2️⃣ Routers Inventory (Data Model)

```python
routers = [
    {
        "name": "R1",
        "device_type": "cisco_ios",
        "host": "192.168.54.142",
        "username": "admin",
        "password": "admin123",
        "secret": "admin123",
        "interface": "Ethernet1/0",
        "ip": "10.10.10.1",
        "mask": "255.255.255.252"
    },
```

### What this is:

* This is your **inventory**, similar to:

  * Ansible inventory
  * NAPALM YAML
* Each dictionary represents **one router**

### Key fields:

| Key                 | Meaning                                 |
| ------------------- | --------------------------------------- |
| `name`              | Logical name (for logging)              |
| `device_type`       | Tells Netmiko how to talk to the device |
| `host`              | Management IP of router                 |
| `username/password` | SSH credentials                         |
| `secret`            | Enable password                         |
| `interface`         | Interface to configure                  |
| `ip/mask`           | IP address to assign                    |

💡 **Important:**
Only these keys are actually required by Netmiko:

* `device_type`
* `host`
* `username`
* `password`
* `secret`

The rest (`interface`, `ip`, `mask`) are **your own variables** used for automation logic.

---

## 3️⃣ Looping Through Routers

```python
for router in routers:
```

* The script runs the same steps for **R1 and R2**
* This is why automation scales:

  * 2 routers → works
  * 200 routers → same code

---

## 4️⃣ Connecting to the Router

```python
conn = ConnectHandler(**router)
conn.enable()
```

### What happens:

1. `ConnectHandler(**router)`

   * Python expands the dictionary into arguments
   * Netmiko:

     * Opens an SSH session
     * Detects Cisco IOS prompt
2. `conn.enable()`

   * Enters **privileged EXEC mode (`#`)**
   * Uses the `secret` password

Equivalent manual CLI:

```text
ssh admin@192.168.54.142
enable
```

---

## 5️⃣ Configuration Commands (Network Automation Core)

```python
config_commands = [
    f"interface {router['interface']}",
    f"ip address {router['ip']} {router['mask']}",
    "no shutdown",
    "exit"
]
```

### What this config does (on each router):

#### On R1:

```text
interface Ethernet1/0
 ip address 10.10.10.1 255.255.255.252
 no shutdown
```

#### On R2:

```text
interface Ethernet1/0
 ip address 10.10.10.2 255.255.255.252
 no shutdown
```

### Networking Meaning:

* Assigns IP addresses to interfaces
* Uses a `/30` subnet:

  * Network: `10.10.10.0/30`
  * Usable IPs:

    * R1 → `10.10.10.1`
    * R2 → `10.10.10.2`
* Brings the interface **UP**

This effectively creates a **point-to-point link** between R1 and R2.

---

## 6️⃣ Sending Config to Device

```python
output = conn.send_config_set(config_commands)
```

### What Netmiko does:

* Enters:

  ```text
  configure terminal
  ```
* Sends each command line-by-line
* Exits config mode automatically

Equivalent manual CLI:

```text
conf t
interface Ethernet1/0
ip address 10.10.10.1 255.255.255.252
no shutdown
exit
```

---

## 7️⃣ Saving the Configuration

```python
conn.save_config()
```

### What this does:

* Executes:

  ```text
  write memory
  ```

  or

  ```text
  copy running-config startup-config
  ```

(depending on IOS)

📌 **Very important**

* Without this, config is lost after reload

---

## 8️⃣ Verification Step

```python
verify = conn.send_command("show ip interface brief")
```

### Why this is good practice:

* Confirms:

  * Interface is **up/up**
  * IP address is applied
* Automation should always:

  1. Configure
  2. Verify

Expected output:

```text
Interface       IP-Address      OK? Method Status  Protocol
Ethernet1/0     10.10.10.1      YES manual up      up
```

---

## 9️⃣ Disconnecting

```python
conn.disconnect()
```

* Closes SSH session cleanly
* Prevents resource leaks

---

## 🔁 What This Script Achieves (Big Picture)

### From a networking view:

✔ Creates a routed link between R1 and R2
✔ Assigns IP addresses
✔ Brings interfaces up
✔ Saves configuration

### From an automation view:

✔ Uses inventory-driven automation
✔ Reusable & scalable
✔ No manual CLI typing
✔ Same code works for many devices

````python
from netmiko import ConnectHandler

routers = [
    {
        "name": "R1",
        "connection": {
            "device_type": "cisco_ios",
            "host": "192.168.54.142",
            "username": "admin",
            "password": "admin123",
            "secret": "admin123",
        },
        "interface": "Ethernet1/0",
        "ip": "10.10.10.1",
        "mask": "255.255.255.252"
    },
    {
        "name": "R2",
        "connection": {
            "device_type": "cisco_ios",
            "host": "192.168.54.143",
            "username": "admin",
            "password": "admin123",
            "secret": "admin123",
        },
        "interface": "Ethernet1/0",
        "ip": "10.10.10.2",
        "mask": "255.255.255.252"
    }
]

for router in routers:
    print(f"\n🔧 Connecting to {router['name']} ({router['connection']['host']})")

    conn = ConnectHandler(**router["connection"])
    conn.enable()

    config_commands = [
        f"interface {router['interface']}",
        f"ip address {router['ip']} {router['mask']}",
        "no shutdown",
        "exit"
    ]

    output = conn.send_config_set(config_commands)
    print(output)

    # Save config
    conn.save_config()

    # Verify
    verify = conn.send_command("show ip interface brief")
    print("\n📊 Verification:")
    print(verify)

    conn.disconnect()
````
