# ✅ What Are Fixtures in Python (Pytest)?

In **Pytest**, a *fixture* is a reusable function that **sets up the environment or data needed for a test before it runs**, and optionally **cleans up after the test**.

Think of a fixture as:

> **“A function that prepares something your test needs.”**

***

# 🔍 How Fixtures Work (Based on Diogo’s explanation)

From your call with [Diogo Oliveira](https://www.office.com/search?q=Diogo+Oliveira\&EntityRepresentationId=da2c7871-2c97-4966-8f71-6fe98006d3ce):

*   A test **requests** a fixture by naming it as a parameter.
*   Pytest **runs the fixture function first**, gets the returned value, and **passes it into the test** as a variable.
*   All code inside the fixture runs **before** the test begins.
*   When multiple fixtures are listed as parameters, **each one is called and its results injected**.    [\[Diogo_no_t...CODE GIT.1 | Video\]](https://microsoft-my.sharepoint.com/personal/v-yaalam_microsoft_com/_layouts/15/viewer.aspx?sourcedoc={82aed413-d153-4f3b-b843-5542ca6faa44})

Example in plain English:  
If your test needs a router session, traffic generator, or prepared data — a fixture sets that up and hands it to the test function.

***

# 🧪 Simple Example

```python
import pytest

@pytest.fixture
def sample_data():
    return {"name": "Yaswanth", "role": "Automation"}

def test_example(sample_data):
    assert sample_data["role"] == "Automation"
```

What happens?

*   `sample_data()` runs *before* the test.
*   Its return value is injected into `test_example()`.

# 🎯 Why Fixtures Are Useful

According to the internal **Python Test Automation BDD Framework** document:

> **Fixtures are a great way to manage context between steps** and support pytest’s ecosystem.    [\[Python Tes...Framework | PDF\]](https://microsoft.sharepoint.com/teams/CampusIPLibraries/Campus/Community%20Shared%20IP/2020/09/Python%20Test%20Automation%20BDD%20Framework.pdf?web=1)

### Benefits:

*   Avoid repeated setup code
*   Make tests clean and readable
*   Share data between tests safely
*   Control setup **scope** (e.g., per test, per module, per session)

***

# 📐 Fixture Scope

Pytest lets you decide how long a fixture “lives”:

| Scope      | Meaning                          |
| ---------- | -------------------------------- |
| `function` | runs before every test (default) |
| `module`   | runs once per file               |
| `class`    | runs once per test class         |
| `session`  | runs once per test run           |

This was also mentioned in Diogo’s explanation: a fixture with **module scope runs once per module**.    [\[Diogo_no_t...CODE GIT.1 | Video\]](https://microsoft-my.sharepoint.com/personal/v-yaalam_microsoft_com/_layouts/15/viewer.aspx?sourcedoc={82aed413-d153-4f3b-b843-5542ca6faa44})

# 🧩 Autouse Fixtures

Sometimes you want a fixture to run **without being explicitly requested**.  
Use `autouse=True`.

```python
@pytest.fixture(autouse=True)
def auto_setup():
    print("Runs before every test automatically")
```

# 📁 Conftest.py

In large automation frameworks (like those used by your team), fixtures are stored in **conftest.py**, making them available across many test files.
